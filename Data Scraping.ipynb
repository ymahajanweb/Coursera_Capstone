Get the data from wikipedia in a table

```python
import requests
import pandas as pd
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
My_table = soup.find('table',{'class':'wikitable sortable'})
```

Populate the dataframe from the table

```python
Neighbourhood = []
#read the table into an array
for row in My_table.find_all('tr'):
    col = row.findAll('td')
    if len(col) > 0:
        postcode = col[0].string
        county = col[1].string
        borough = col[2].string
        Neighbourhood.append([t.text.strip() for t in row.find_all('td')])
        
df = pd.DataFrame([])
#populate the dataframe from array
df = pd.DataFrame(Neighbourhood, columns=['PostalCode', 'Borough', 'Neighbourhood'])
#Check the data in the dataframe
df.head()
```

Delete rows with 'Not assigned' borough

```python
 # Get names of indexes for which column 'Borough is Not assigned'
indexNames = df[ df['Borough'] == 'Not assigned' ].index
    
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
df.head()   
```

Group the neighnourhoods by postal code and borough and transform them to a comma seperated list. Drop the duplicates

```python
df['Neighbourhood'] = df.groupby(['PostalCode', 'Borough'])['Neighbourhood'].transform(lambda x: ', '.join(set(x.dropna())))
df.drop_duplicates()
```
